# Model Building & Evaluation
## Q8: Train a Machine Learning model (RandomForestClassifier)

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

from sklearn.model_selection import RandomizedSearchCV
```

```python
RF = RandomForestClassifier()
LR = LogisticRegression()
```

### Random Forest

```python
RF.fit(X_train_sm,y_train_sm)
```

```python
y_pred = RF.predict(x_test)
y_pred
```

```python
#show accuracy score 
accuracy_score(y_test,y_pred)
```

```python
#show confusion matrix
confusion_matrix(y_test, y_pred)
```

### using RandomSearchCV for using the best parameter to train model

```python
#using RandomSearchCV
n_estimator = [20, 30, 40, 50, 60, 70 ,100, 200, 300]
max_features = [0.2, 0.3, 0.5]
max_depth = [2,4,6,8,None]
max_sample = [0.25,0.5,0.75,1.0]
```

```python
param_grid = {'n_estimators': n_estimator,
              'max_features': max_features,
              'max_depth': max_depth,
              'max_samples':max_sample,
             }
print(param_grid)
```

```python
RF_grid = RandomizedSearchCV(estimator = RF, 
                             param_distributions = param_grid, 
                             cv = 5, 
                             verbose=2, 
                             n_jobs = -1)
```

```python
RF_grid.fit(X_train_sm,y_train_sm)
```

```python
y_predict = RF_grid.predict(x_test)
y_predict
```

```python
#show accuracy score 
accuracy_score(y_test,y_predict)
```

```python
#show confusion matrix
confusion_matrix(y_test, y_predict)
```

```python
RF_grid.best_params_
```
